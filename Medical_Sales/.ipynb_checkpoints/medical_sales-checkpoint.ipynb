```python
import numpy as np
import pandas as pd
```

```python
file_path = "/Users/arjun/Downloads/Pharmacy_OLTP_SLStyle_18Months.csv"
sales = pd.read_csv(file_path)
sales.head()
```

```python
print("Total rows:", len(sales))
```

```python
print("Shape:", sales.shape)
cats = sales.select_dtypes(include='object').columns
sankey_ready = {col: sales[col].nunique() for col in cats}
print("Categories for Sankey:")
for col, n in sankey_ready.items():
    print(f"  {col}: {n} {'✅' if n >= 3 else '⚠️'}")

# Sankey viable if any chain >=3 cols with 3+ cats each
viable_chains = sum(1 for chain in [['Pharmacy','Branch','Province'], ['Branch','Province','Payment_method']] 
                    if all(sankey_ready.get(c,0) >=3 for c in chain))
print(f"\nViable Sankey flows: {viable_chains}/2 ✅")
```
